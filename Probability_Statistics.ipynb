Simulation of a trial (a single sequence of n experiments)

```python
# Simulation of a trial (a single sequence of n experiments)
import numpy as np
from scipy.stats import binom

n = 10 # number of experiments in a trial
p = 0.5 # probability at each experiment

np.random.binomial(10, 0.5)
```

```python
# Probability of getting 6 heads with 10 coin flips by running 10000 simulations
trials = 10_000
prob_6 = sum([1 for i in np.random.binomial(n, p, size=trials) if i == 6]) / trials
print(f'The probability of 6 heads in the simulation is: {prob_6}')
print(f'The probability of 6 heads in theory is: {binom.pmf(6, 10, 0.5)}')
```

Simulation of 1000 trials of flipping 10 coins.

If random value >= 0.5 then we accept that a coin fell heads.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def run_binom(trials, n, p):
    heads = []
    for i in range(trials):
        tosses = [np.random.random() for i in range(n)]
        heads.append(len([i for i in tosses if i >= p]))
    return heads
```

```python
trials = 1000 # number of trials
n = 10 # number of experiments in a trial
p = 0.5 # probability at each experiment

# Simulate binomial distribution, tossing a coin 10 times and repeating the trial 1000 times
heads = run_binom(trials, n, p)
print(heads)

# Plot the simulated binomial distribution
fig, ax = plt.subplots(figsize=(14, 7))
bins = np.arange(min(heads) - 0.5, max(heads) + 1, 1)
ax = sns.histplot(heads, kde=True, bins=bins, label='Simulated results')
ax.set_xlabel('Number of heads', fontsize=16)
ax.set_ylabel('Frequency', fontsize=16)
ax.legend()
```

```python
from scipy.stats import binom

fig, ax = plt.subplots(figsize=(14, 7))
bins = np.arange(min(heads) - 0.5, max(heads) + 1, 1)
ax = sns.distplot(heads, kde=True, bins=bins, label='Simulated results')
ax.set_xlabel('Number of heads', fontsize=16)
ax.set_ylabel('Frequency', fontsize=16)

# Plot the actual binomial distribution
x = range(0, max(heads) + 1)
ax.plot(x, binom.pmf(x, n, p), 'ro', label='Actual binomial distribution')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='r', lw=5, alpha=0.5)
plt.legend()
plt.show()
```

### Real World Applications of the Binomial Distribution
1. The typical call center employee completes on average 50 calls per day.
2. The probability of a conversion (purchase) for each call is 4%.
3. The average revenue to your company for each conversion is \$100.
4. The call center you are analyzing has 100 employees.
5. Each employee is paid \$200 per day of work.

```python
# Call center simulation
import numpy as np
import seaborn as sns
from scipy.stats import binom

# Number of employees to simulate (trials)
employees = 100

# Cost per employee
wage = 200

# Number of independent calls per employee (experiments)
n = 50

# Probability of success for each call
p = 0.04

# Revenue per call
revenue = 100

# Binomial random variables of call center employees
conversions = np.random.binomial(n, p, size=employees)
print('Conversions for 100 employees per one day (4% chance of success): ', conversions)
```

```python
ax = plt.subplots(figsize=(14,7))

# Plot the simulated result
bins = np.arange(conversions.min() - 0.5, conversions.max() + 1, 1)
ax = sns.histplot(conversions, stat='probability', bins=bins, label='Simulated result (4% chance of success)')

# Plot the actual binomial distribution
x = range(0, max(conversions) + 1, 1)
y = binom.pmf(range(0, max(conversions) + 1, 1), n, p)
ax.plot(x, y, 'ro', label='Actual binomial distribution')
ax.vlines(x, 0, y, colors='red', lw=5, alpha=0.5)
ax.legend()
```

These are results for just one randomly generetad day:

```python
# results for just one randomly generetad day

print(f'Average Conversions per Employee: {np.mean(conversions):.2f}\n\
Standart deviation of conversions per Employee: {np.std(conversions):.2f}\n\
Total Conversions: {np.sum(conversions)}\n\
Total revenues: {np.sum(conversions) * revenue}\n\
Total Expenses: {employees * wage}\n\
Total Profits \ Losses: {np.sum(conversions) * revenue - employees * wage}')
```

```python
# Simulation of 1000 days of work of the call center

# Number of days to simulate
sims = 1000

# Simulation of 1000 days of work of the call center
sim_conversions = [np.sum(np.random.binomial(n, p, size=employees)) for i in range(sims)]
sim_profit = np.array(sim_conversions) * revenue - employees * wage
```

```python
ax = plt.subplots(figsize=(14,7))

bins = np.arange(min(sim_profit) - 250, max(sim_profit) + 500, 500)
ax = sns.histplot(sim_profit, stat='probability', bins=bins, kde=True, label='Original call center simulation results')
```

Attempt to improve results of the call center

```python
# Call Center Simulation (Higher Conversion Rate)

# Number of employees to simulate (trials)
employees = 100

# Cost per employee
wage = 200

# Number of independent calls per employee (trial)
n = 55

# Probability of success for each call
p = 0.05

# Revenue per call
revenue = 100

# Binomial random variables of call center employees
conversions_up = np.random.binomial(n, p, size=employees)
conversions_up
```

```python
ax = plt.subplots(figsize=(14,7))

# Plot the improved simulation results
bins = np.arange(conversions_up.min() - 0.5, conversions_up.max() + 1, 1)
ax = sns.histplot(conversions_up, bins=bins, stat='probability', label='The improved simulated results (5% chance of success)')

# Plot the actual binomial distribution
x = range(0, max(conversions_up) + 1, 1)
y = binom.pmf(range(0, max(conversions_up) + 1, 1), n, p)
ax.plot(x, y, 'ro', label='Actual binomial distribution')
ax.vlines(x, 0, y, colors='r', lw=5, alpha=0.5)
ax.legend()
```

```python
# Simulate 1000 days for our improved call center

# Number of days to simulate
sims = 1000

# Simulation of 1000 days of work of improved call center
sim_conversions_up = [np.sum(np.random.binomial(n, p, size=employees)) for i in range(sims)]
sim_profits_up = np.array(sim_conversions_up) * revenue - employees * wage
```

```python
# Plot and save the results with increased parameters as a histogram
fig, ax = plt.subplots(figsize=(14, 7))
bins_up = np.arange(sim_profits_up.min() - 250, sim_profits_up.max() + 500, 500)
ax = sns.histplot(sim_profits_up, stat='probability', bins=bins_up, kde=True, label='improved call center simulation results', color='red', legend=True)
ax.legend()
```

```python
# Comparision graphics
fig, ax = plt.subplots(figsize=(14, 7))

bins = np.arange(min(sim_profit) - 250, max(sim_profit) + 500, 500)
ax = sns.histplot(sim_profit, stat='probability', bins=bins, kde=True, label='Original call center simulation results')

bins_up = np.arange(sim_profits_up.min() - 250, sim_profits_up.max() + 500, 500)
ax = sns.histplot(sim_profits_up, stat='probability', bins=bins_up, kde=True, label='improved call center simulation results', color='red')

ax.legend()
```
