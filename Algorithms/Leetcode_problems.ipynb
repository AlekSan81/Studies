## Roman to Integer

Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M.

Symbol       Value
I             1
V             5
X             10
L             50
C             100
D             500
M             1000
For example, 2 is written as II in Roman numeral, just two one's added together. 12 is written as XII, which is simply X + II. The number 27 is written as XXVII, which is XX + V + II.

Roman numerals are usually written largest to smallest from left to right. However, the numeral for four is not IIII. Instead, the number four is written as IV. Because the one is before the five we subtract it making four. The same principle applies to the number nine, which is written as IX. There are six instances where subtraction is used:

I can be placed before V (5) and X (10) to make 4 and 9. 
X can be placed before L (50) and C (100) to make 40 and 90. 
C can be placed before D (500) and M (1000) to make 400 and 900.
Given a roman numeral, convert it to an integer.

```python
s = 'DCCLIV'

# Ищем числа с вычитанием
num_dict = {}
if 'IV' in s:
    num_dict['IV'] = 4
    s = s.replace('IV', '')
if 'IX' in s:
    num_dict['IX'] = 9
    s = s.replace('IX', '')
if 'XL' in s:
    num_dict['XL'] = 40
    s = s.replace('XL', '')
if 'XC' in s:
    num_dict['XC'] = 90
    s = s.replace('XC', '')
if 'CD' in s:
    num_dict['CD'] = 400
    s = s.replace('CD', '')
if 'CM' in s:
    num_dict['CM'] = 900
    s = s.replace('CM', '')
    
# меняем римские цифры на обычные   
rom_sym_list = list(s)
for i in range(len(rom_sym_list)):
    if rom_sym_list[i] == 'I':
        rom_sym_list[i] = 1
    elif rom_sym_list[i] == 'V':
        rom_sym_list[i] = 5
    elif rom_sym_list[i] == 'X':
        rom_sym_list[i] = 10
    elif rom_sym_list[i] == 'L':
        rom_sym_list[i] = 50
    elif rom_sym_list[i] == 'C':
        rom_sym_list[i] = 100
    elif rom_sym_list[i] == 'D':
        rom_sym_list[i] = 500
    else:
        rom_sym_list[i] = 1000

summa = sum(num_dict.values()) + sum(rom_sym_list)
print(summa)
```

```python
class Solution(object):
    def romanToInt(self, s):
        """
        :type s: str
        :rtype: int
        """
        values = {
        "I": 1,
        "V": 5,
        "X": 10,
        "L": 50,
        "C": 100,
        "D": 500,
        "M": 1000,
        "IV": 4,
        "IX": 9,
        "XL": 40, 
        "XC": 90,
        "CD": 400,
        "CM": 900
}
        i = 0
        sum = 0
        while i < len(s):
            if i+1 < len(s) and s[i:i+2] in values:
                sum = sum + values[s[i:i+2]]
                i+=2
            else:
                sum = sum + values[s[i]]
                i+=1
        return sum
    
q = Solution()
print(q.romanToInt('DCCXLIII'))
```

## Longest Common Prefix

Write a function to find the longest common prefix string amongst an array of strings.

If there is no common prefix, return an empty string "".

```python
class Solution:
    def longestCommonPrefix(self, strs):
              
        prefix = []
        
        for letter_ind in range(len(strs[0])):
            for word in strs[1:]:
                if letter_ind+1 > len(word):
                    return ''.join(prefix)                
                if strs[0][letter_ind] == word[letter_ind]:
                    continue
                else:
                    return ''.join(prefix)
            prefix.append(strs[0][letter_ind]) 
       
        return ''.join(prefix)
        
q = Solution()
q.longestCommonPrefix(["ab", "a"])
```

```python
class Solution:
    def longestCommonPrefix(self, strs) -> str:
        
        if not strs:
            return ""
        
        prefix = strs[0]
        for s in strs[1:]:
            
            # prefix isn't shorter so we can skip this entry
            if s.startswith(prefix):
                continue
            
            count = 0
            for l, r in zip(prefix, s):
                if l != r:
                    break
                count += 1
                
            # no shared characters, which means no common prefix
            if count == 0:
                return ""
                
            prefix = prefix[:count]
            
        return prefix
    
q = Solution()
q.longestCommonPrefix(["ab", "a"])
#q.longestCommonPrefix(["flower","flow","flight"])
#q.longestCommonPrefix(["dog","racecar","car"])
```

**Гипотеза**: при правильной последовательности скобок, всегда будут скобки которые открываются и тут же (следующим символом) закрываются.  

***Алгоритм решения (вымученный, корявый, но работающий)***:  

1. Ищем в строке скобку, которая открывается и тут же закрывается (следующим символом).
2. Удаляем оба этих символа.
3. В новой (уменьшенной) строке повторяем эту операцию, до тех пор пока len(s) == 0. После чего функция возвращает True.  
Если же в какой то момент программы алгоритм не найдет открывающуюся и тут же закрывающуюся скобку, то это значит, что последовательность скобок ошибочна и функция возвращает False.

```python
class Solution:
    def isValid(self, string) -> bool:
        
        while len(string) > 0:
            
            # Ищем в строке скобку, которая открывается и тут же закрывается (следующим символом)    
            if len(string) % 2 != 0:
                return False
            for symbol_index in range(len(string) - 1):
                if (string[symbol_index] == '(' and string[symbol_index + 1] == ')') or\
                   (string[symbol_index] == '[' and string[symbol_index + 1] == ']') or\
                   (string[symbol_index] == '{' and string[symbol_index + 1] == '}'):
                    
                    # Если найдена пара "правильных" скобок, то они удаляются из строки                    
                    if symbol_index + 2 < len(string):
                        string = string[ : symbol_index] + string[symbol_index + 2 : ]
                       
                        # после чего цикл прерывается и с новой "уменьшенной" строкой проводится новая итерация
                        break
                        
                    else:
                        string = string[ : symbol_index]
                        break
                
                # Если после всех удалений в строке и после полного прохождения цикла в строке еще остались символы
                # то это говорить, что скобки в неправильном порядке, функция возвращает False
                elif symbol_index + 2 == len(string):
                    return False      
        
        # Если программа вышла из цикла while, значит строка пустая, все скобки были удалены, т.к.
        # соответствовали главной гипотизе "правильности".
        return True

q = Solution()
print(q.isValid("[({(())}[()])]"))
```

**Гипотеза**: скобки закрываются в обратном порядке их открытию (та которая открылась последней, должна закрываться первой, и та которая открылась первой должна закрываться последней). 

***Алгоритм (победитель)***:  
 
1. Создается "показательный" словарь, где ключ - открывающая скобка, значение - закрывающая скобка. Данный словарь, как образец, он не меняется во время исполнения программы.
2. Создается пустой список, в который при переборе строки будут заносится закрывающие скобки (значения из словаря) согласно открывающей скобки (по ключу словаря). В программе он назван: "контролирующий список".
3. Если при переборе строки попадает закрывающая скобка, то проверяется, были ли перед ней какие-либо открывающие скобки, т.е. "контрольный" список к этому моменту не должен быть пустым. Если же он пустой, функция возвращает False.
4. Программа проверяет последний символ контролирующего списка, он должен совпадать с текущей закрывающей скобкой цикла. Если это не так, то программа возвращает False.
5. После прохождения всего цикла "контрольный" список должен быть пустым, т.е. все открытые скобки были закрыты.

```python
class Solution:
    def isValid(self, s: str) -> bool:
        parentheses = {'(': ')', '[': ']', '{': '}'}
        control = []
    
        for i in s:
            if i in parentheses: # 1
                control.append(parentheses[i])
            elif len(control) == 0: # 2
                return False
            elif control.pop() != i: # 3
                return False

        if len(control) == 0: # 4
            return True
        else:
            return False
		
		# 1 We have to control if i is on the dictionary as a key. if it is, add the value of this key to list.
		
		# 2 it controls if we add paranthesis in step 1 in the previous loop.
		
		# 3 return and remove the last element of list and control it if it is NOT same with i.
		# if the last element is same with i this means we found the pair.
		# if they are not same this means we couldnt find the pair of the parentheses in last index. it breaks the order
		
		# 4 in step 3 we removed all the parentheses that we found their pairs. ( with control.pop() )
		# if the length of control list is 0 this means we found the pairs of all parenthesis
    
q = Solution()
print(q.isValid("[({(())}[()])]"))
```

## Merge Two Sorted Lists
Merge two sorted linked lists and return it as a sorted list. The list should be made by splicing together the nodes of the first two lists.  

Input: l1 = [1,2,4], l2 = [1,3,4]  
Output: [1,1,2,3,4,4]

Гипотеза: Проверяем два списка, есть ли значения.
Алгоритм:
1. Создать словарь

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeTwoLists(self, l1, l2):
           
            head =  ListNode(0)
            cur_node = head
            while(l1 and l2):
                if(l1.val<l2.val):
                    cur_node.next = l1
                    l1 = l1.next
                else:
                    cur_node.next = l2
                    l2 = l2.next
                cur_node = cur_node.next
                
            if(l1):
                cur_node.next = l1
            elif(l2):
                cur_node.next = l2

            return head.next
    

l1 = ListNode([1,2,4])
l2 = ListNode([1,3,4])

q = Solution()
q.mergeTwoLists(l1, l2)
```

```python
class Solution:
    def removeElement(self, nums, val):
        # Logic is we move all vals to the end of list.
        # lastIndex will point to the index from which all elements to its right will
        # be val only.
        # We initialise it with len(nums) - 1 at the start
        lastIndex = len(nums) - 1
        # index used to loop through list
        index = 0
        # We loop as long as index <= lastIndex. Because, elements to the right
        # of lastIndex will be val only
        while index <= lastIndex:
            # If nums[index] is val, we swap it with the element present at element
            # at nums[lastIndex]
            # We won't move index to its next position. Because there is a possibility
            # that element at lastIndex is val before placing val in it.
            if nums[index] == val:
                nums[index], nums[lastIndex] = nums[lastIndex], nums[index]
                # We move lastIndex one step to its left. Because, we already used
                # that spot to place a val from index position
                lastIndex -= 1
            # if nums[index] is not val, we increment index
            else: index += 1
        # Elements from index 0 to lastIndex won't have val. So we return lastIndex + 1
        # Why we return lastIndex + 1 is, lastIndex is 0 based and we need a value
        # based on 1 based because we need number of elements that are not val.    
        return lastIndex + 1

nums = [0,1,2,2,3,0,4,2]
val = 2
inst = Solution()
inst.removeElement(nums, val)
```
