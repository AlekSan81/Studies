#### Bubble sort

```python
# Bubble sort

lst = [37, 5, 7, 3, 2, 1, 9, 14, 4, 1]

for i in range(len(lst)):
    for j in range(len(lst) - i - 1):
        if lst[j] > lst[j+1]:
            lst[j], lst[j+1] = lst[j+1], lst[j]

print(lst)
```

```python
# Adding swap check to stop the algorithm, if a swap has not been detected in last iteration
# Plus putting the sort algorithm in a function

def bubble_sort(lst):
    for i in range(len(lst)):
        swap = False
        for j in range(len(lst) - 1 - i):
            if lst[j] > lst[j + 1]:
                lst[j], lst[j + 1] = lst[j + 1], lst[j]
                swap = True

        if swap == False:
            return lst


lst = [37, 5, 7, 3, 2, 1, 9, 14, 4, 1]
bubble_sort(lst)
```
