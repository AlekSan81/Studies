Simulation of 1000 trials of flipping 10 coins.

If random value >= 0.5 then we accept that a coin fell heads.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def run_binom(trials, n, p):
    heads = []
    for i in range(trials):
        tosses = [np.random.random() for i in range(n)]
        heads.append(len([i for i in tosses if i >= p]))
    return heads
```

```python
trials = 1000 # number of trials
n = 10 # number of experiments in a trial
p = 0.5 # probability at each experiment

# Simulate binomial distribution, tossing a coin 10 times and repeating the trial 1000 times
heads = run_binom(trials, n, p)
print(heads)

# Plot the simulated binomial distribution
fig, ax = plt.subplots(figsize=(14, 7))
bins = np.arange(min(heads) - 0.5, max(heads) + 1, 1)
ax = sns.histplot(heads, kde=True, bins=bins, label='Simulated results')
ax.set_xlabel('Number of heads', fontsize=16)
ax.set_ylabel('Frequency', fontsize=16)
ax.legend()
```

```python
from scipy.stats import binom

fig, ax = plt.subplots(figsize=(14, 7))
bins = np.arange(min(heads) - 0.5, max(heads) + 1, 1)
ax = sns.distplot(heads, kde=True, bins=bins, label='Simulated results')
ax.set_xlabel('Number of heads', fontsize=16)
ax.set_ylabel('Frequency', fontsize=16)

# Plot the actual binomial distribution
x = range(0, max(heads) + 1)
ax.plot(x, binom.pmf(x, n, p), 'ro', label='Actual binomial distribution')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='r', lw=5, alpha=0.5)
plt.legend()
plt.show()
```

### Real World Applications of the Binomial Distribution
1. The typical call center employee completes on average 50 calls per day.
2. The probability of a conversion (purchase) for each call is 4%.
3. The average revenue to your company for each conversion is \$100.
4. The call center you are analyzing has 100 employees.
5. Each employee is paid \$200 per day of work.

```python
# Call center simulation
import numpy as np
import seaborn as sns
from scipy.stats import binom

# Number of employees to simulate (trials)
employees = 100

# Cost per employee
wage = 200

# Number of independent calls per employee (experiments)
n = 50

# Probability of success for each call
p = 0.04

# Revenue per call
revenue = 100

# Binomial random variables of call center employees
conversions = np.random.binomial(n, p, size=employees)
print('Conversions for 100 employees per one day (4% chance of success): ', conversions)
```

```python
ax = plt.subplots(figsize=(14,7))

# Plot the simulated result
bins = np.arange(conversions.min() - 0.5, conversions.max() + 1, 1)
ax = sns.histplot(conversions, stat='probability', bins=bins, label='Simulated result (4% chance of success)')

# Plot the actual binomial distribution
x = range(0, max(conversions) + 1, 1)
y = binom.pmf(range(0, max(conversions) + 1, 1), n, p)
ax.plot(x, y, 'ro', label='Actual binomial distribution')
ax.vlines(x, 0, y, colors='red', lw=5, alpha=0.5)
ax.legend()
```

These are results for just one randomly generetad day:

```python
# results for just one randomly generetad day

print(f'Average Conversions per Employee: {np.mean(conversions):.2f}\n\
Standart deviation of conversions per Employee: {np.std(conversions):.2f}\n\
Total Conversions: {np.sum(conversions)}\n\
Total revenues: {np.sum(conversions) * revenue}\n\
Total Expenses: {employees * wage}\n\
Total Profits \ Losses: {np.sum(conversions) * revenue - employees * wage}')
```

```python
# Simulation of 1000 days of work of the call center

# Number of days to simulate
sims = 1000

# Simulation of 1000 days of work of the call center
sim_conversions = [np.sum(np.random.binomial(n, p, size=employees)) for i in range(sims)]
sim_profit = np.array(sim_conversions) * revenue - employees * wage
```

```python
ax = plt.subplots(figsize=(14,7))

bins = np.arange(min(sim_profit) - 250, max(sim_profit) + 500, 500)
ax = sns.histplot(sim_profit, stat='probability', bins=bins, kde=True, label='Original call center simulation results')
```

### The Poisson's discreet probability distribution

```python
import math
from scipy.stats import binom
```

```python
# The Poisson's formula of probability distribution
def poisson_probability(i, lambda_poisson):
    return (lambda_poisson**i / (math.factorial(i) * math.exp(lambda_poisson)))
```

#### Problem 1
A taxi dispatcher has 30 calls per hour. Find:
1. probability of 2 or 3 calls per minute
2. probability of at least 1 call per 5 minute

```python
# Number of calls per hour
n = 30

# transforming hours to minutes to get the average number of calls per minute
p = 1/60

# The Poisson's lambda (the average for 1 minute)
lambda_poisson = n * p
print(f'The lambda is: {lambda_poisson:.4f}. It is the average number of calls per minute.')

calls_probabilities = []
for i in range(2,4):
    current_probability = poisson_probability(i, lambda_poisson)
    calls_probabilities.append(current_probability)
    print(f'The probability of {i} call(s) per minute is {current_probability:.4f}.')
print(f'The overall probability is {sum(calls_probabilities):.4f}.')
```

```python
# Verification by binomial function
print(binom.pmf([2, 3], n, p))
sum(binom.pmf([2, 3], n, p))
```

```python
# transforming hours to minutes to get the average number of calls per 5 minute
p = 5/60

# The Poisson's lambda (the average for 5 minute)
lambda_poisson = n * p
print(f'The lambda is: {lambda_poisson:.4f}. It is the average number of calls per 5 minutues.')

calls_probabilities = []
for i in range(1):
    current_probability = poisson_probability(i, lambda_poisson)
    calls_probabilities.append(current_probability)
    print(f'The probability of {i} call(s) per 5 minute is {current_probability:.4f}.')
print(f'The probability of at least 1 call per 5 minute is {1 - sum(calls_probabilities):.4f}.')
```

```python
# Verification by binomial function
print(binom.pmf([0], n, p))
1 - sum(binom.pmf([0], n, p))
```

#### Problem 2
The average number of cars, checked at customs, is 3 per hour (lambda = 3).
Find the probability of checking 7 to 10 cars in 2 hours.

```python
# The lambda for 2 hours
lambda_poisson = 3 * 2

number_cars = []
cars_probabilities = []
for i in range(7,11):
    current_probability = poisson_probability(i, lambda_poisson)
    number_cars.append(i)
    cars_probabilities.append(current_probability)
    print(f'The probability of checking {i} cars is: {current_probability:.4f}.')

print(f'The overall probability of checking {number_cars[0]} to {number_cars[-1]} cars is: {sum(cars_probabilities):.4f}.')
number_cars
```

```python
# Verification by binomial function
print(binom.pmf(number_cars, 72, 2/24))
sum(binom.pmf(number_cars, 72, 2/24))
```

Find the probability of only 1 car to be checked in half hour

```python
# The lambda for half an hour
lambda_poisson = 3 * 1/2

car_probability = poisson_probability(1, lambda_poisson)
print(f'The probability of only one car to be checked in half an hour is: {car_probability:.4f}')
```

## Hypergeometric distribution

```python
from scipy.stats import hypergeom
import matplotlib.pyplot as plt
```

http://mathprofi.ru/gipergeometricheskoe_raspredelenie_veroyatnostei.html

The bin has 4 black and 6 white balls, which are randomly extracted from the bin wih no return.  
Estimate:
1. The probability mass function of the random variable X.
2. The Expected value and variance
3. Cumulative distribution function
4. The probability of drawing at least 2 black balls
5. $P(|X - E| < sd)$

### The probability mass function of the random variable X.

```python
# Use of frozen distribution
# The bin has N=10 balls, M=4 black balls, we randomly extract n=3 balls
[N, M, n] = [10, 4, 3]
rv = hypergeom(N, M, n)
x = np.arange(0, M)
pmf_balls = rv.pmf(x)
pmf_balls
```

```python
# Use of hypergeom methods directly, instead of using a frozen distribution
'''
hypergeom.pmf(k, N, M, n)
k - target value or range of values
N - overall number of items in the basket
M - number of items with a specific feature
n - number of extracted items
'''
hypergeom.pmf(range(0, M), 10, 4, 3)
```

```python
fig = plt.figure(figsize=(14,7))
ax = fig.add_subplot()
ax.plot(x, pmf_balls, 'bo')
ax.plot(x, pmf_balls)
ax.vlines(x, 0, pmf_balls, lw=5, alpha=0.5)
ax.set_xlabel('Number of black balls in the group of extracted balls')
ax.set_ylabel('hypergeom PMF')
```

### Expected value and variance (+ standart deviation)

```python
expect = hypergeom.expect(args=(N, M, n))
# or
expect = np.average(x, weights=pmf_balls)
print(f'Expected value: {expect:.4f}')

var = hypergeom.var(N, M, n)
print(f'Variance: {var}')

sd = hypergeom.std(N, M, n)
print(f'Standart deviation: {sd:.4f}')
```

### Cumulative distribution function

```python
x = range(0, M)
cdf_balls = hypergeom.cdf(x, N, M, n)
print(f'Cumulative distribution: {cdf_balls}')
```

```python
fig = plt.figure(figsize=(14,7))
ax = fig.add_subplot()
ax.plot(x, cdf_balls)
ax.plot(x, cdf_balls, 'bo')
ax.set_xlabel('Number of black balls in the group of extracted balls')
ax.set_ylabel('Cumulative probability')
ax.grid()
```

### Estimating $P$
$P (X >= 2) = F(+\infty) - F(2)$

```python
round(cdf_balls[3] - cdf_balls[1], 4)
```

$P(|X - E(X)| < sd) = F(E(X) + sd) - F(E(X) - sd)$

```python
round(cdf_balls[1] - cdf_balls[0], 4)
```
